# Find average rating of movies year wise

```python
data.columns
```

```python
avg_rating_per_year=data.groupby("Year")['Rating'].mean().sort_values(ascending = False)
avg_rating_per_year
```

```python
plt.figure(figsize=(10, 6)) 
plt.plot(avg_rating_per_year.index, avg_rating_per_year.values, 
         marker='o', linestyle='-', color='darkblue', 
         markersize=8, markerfacecolor='lightblue', markeredgecolor='darkblue',)
plt.title("Average Movie Rating Over the Years", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.grid(True, color='lightgray', linestyle='--', linewidth=0.5)
plt.xticks(rotation=45) 
plt.axhline(y=data['Rating'].mean(), color='red', linestyle='--', label='Overall Average')
plt.legend()
plt.show()
```

# Counting movies per year



```python

movies_per_year = data['Year'].value_counts().reset_index()
movies_per_year.columns = ['Year', 'Count']
movies_per_year = movies_per_year.sort_values('Year')
plt.figure(figsize=(10, 6))
sns.barplot(data=movies_per_year, x='Year', y='Count')
plt.title('Number of Movies Released Each Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.xticks(rotation=45)
plt.show()

```

# Scatter plot of rating vs revenue

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Rating', y='Revenue (Millions)', hue='Year', palette='deep')
plt.title('Rating vs Revenue (Millions)', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Revenue (Millions)', fontsize=12)
plt.show()
```

```python
genre_counts = data['Genre'].str.split(',').explode().value_counts()
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index,startangle=160, colors=sns.color_palette('pastel'))
plt.title('Proportion of Movies by Genre', fontsize=16)
plt.show()
```
